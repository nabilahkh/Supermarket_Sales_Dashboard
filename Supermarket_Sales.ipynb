```python
# Changing the data type of the 'Order Date' and 'Ship Date' columns from object to datetime

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

Counting the number and percentage of missing values in the columns of the 'Supermarket' dataset, then displaying them back in a new DataFrame containing such information, such as column names, the count of missing values, and the percentage of missing values in each column.

```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values/len(df)) * 100
missing_info = pd.DataFrame({
    'Column' : missing_values.index,
    'Missing Values' : missing_values.values,
    'Missing Percentage' : missing_percentage.values
})
print(missing_info)
```
